## Handling the missing values

#### The dataset contains float64, int64 and object types. Columns 2, 7, 10 and 14 contain numeric values (of types float64, float64, int64 and int64 respectively) and others contain non-numeric values. 
#### The dataset has missing values "?", now let's replace these missing value question marks with NaN


```python
# Import numpy
import numpy as np
```

```python
# Replace the '?'s with NaN
df = df.replace('?', np.nan)
```

```python
#check missing values again
df.tail(20)
```

#### Impute the missing values with the mean for numerical variables

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

#### Let's replace missing values for columns with objective types

```python
# Iterate over each column of cc_apps
for col in df.columns:
    # Check if the column is of object type
    if df[col].dtypes == 'object':
        # Impute with the most frequent value
        df = df.fillna(df[col].value_counts().index[0])
```

```python
# Check missing values again
print(df.isnull().sum())
```

## Preprocessing the data

Convert the non-numeric data into numeric.
Remove unimportant variables.
Split the data into train and test sets.
Scale the feature values to a uniform range.

#### Convert the non-numeric data into numeric

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le=LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in df.columns.to_numpy():
    # Compare if the dtype is object
    if df[col].dtypes=='object':
    # Use LabelEncoder to do the numeric transformation
        df[col]=le.fit_transform(df[col])
```

#### Remove DriversLicense and ZipCode features from the dataset because they are not important for predicting credit card approvals

```python
# Drop the features DriversLicense and ZipCode  and convert the DataFrame to a NumPy array
df = df.drop([11, 13], axis=1)
df = df.to_numpy()
```

#### Split the data into train and test sets

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Segregate features and labels into separate variables
X,y = df[:,0:13] , df[:,13]

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=50)
```

#### Scale the feature values

The credit score of a person is their creditworthiness based on their credit history. The higher this number, the more financially trustworthy a person is considered to be. So, a CreditScore of 1 is the highest since we're rescaling all the values to the range of 0-1.

```python
# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
# Instantiate MinMaxScaler and use it to rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

## Fitting a logistic regression model to the train set

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Instantiate a LogisticRegression classifier with default parameter values
logreg = LogisticRegression()

# Fit logreg to the train set
logreg.fit(rescaledX_train,y_train)
```

#### Predict and evaluate performance

```python
# Import confusion_matrix
from sklearn.metrics import confusion_matrix

# Use logreg to predict instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression: ", logreg.score(rescaledX_test,y_test))

# Print the confusion matrix of the logreg model
confusion_matrix(y_test,y_pred)
```

The model was pretty good. It was able to yield an accuracy score of almost 88%

For the confusion matrix, the first element of the of the first row of the confusion matrix denotes the true negatives meaning the number of negative instances (denied applications) predicted by the model correctly. And the last element of the second row of the confusion matrix denotes the true positives meaning the number of positive instances (approved applications) predicted by the model correctly.
